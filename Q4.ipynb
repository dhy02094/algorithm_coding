```python
# Q4 answer template

def solution(k, a, b):
    answer = 0
    a.sort()                # [1,2,3,4,5]
    b.sort(reverse=True)    # [6,6,5,5,5]
    
    for i in range(k):      # a   k b .
        if a[i] < b[i]:
            a[i] = b[i]
    answer = sum(a)         # ~
    return answer

k = 3
a = [1,2,5,4,3]
b = [5,5,6,6,5]
answer = solution(k, a, b)
print(answer)
```

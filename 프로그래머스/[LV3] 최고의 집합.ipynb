```python
def solution(n, s):
    answer = []
    
    a = s//n
    if a < 1 :
        answer = [-1]
    else :
        for i in range(n):
            answer.append(a)
            b = s % n
            
        for j in range(b):
            answer[j] += 1
        answer.sort()               
                                
    return answer
```

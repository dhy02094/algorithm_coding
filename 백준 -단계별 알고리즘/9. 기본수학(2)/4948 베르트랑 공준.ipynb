```python
def prime_list(n):
    prime = [True] * (n+1)
    
    m = int(n**0.5)
    
    if n <= 3 :
        prime[0] = False
        return [i+1 for i in range(n) if prime[i] == True]
    else:
        for i in range(2,m+1):
            if prime[i] == True:
                for j in range(i+i, n+1, i):
                    prime[j] = False
        return [i for i in range(2,n+1) if prime[i] == True]

while 1:
    a = int(input())
    if a == 0 :
        break
    print(len(prime_list(2*a)) - len(prime_list(a)))

        
            
```

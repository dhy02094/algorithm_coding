```python
a = int(input())
b = int(input())

prime = []

for i in range(a,b+1):
    cnt = 0
    if i ==1 :
        continue
    for j in range(2,b+1):
        if i % j == 0:
            cnt += 1
        if cnt == 2:
            break       
    if cnt == 1 :
        prime.append(i)
        
if prime == []:
    print(-1)
else:
    print(sum(prime))
    print(min(prime))

        
```

```python
# 시간초과

a,b = map(int, input().split())

for i in range(a,b+1):
    cnt = 0
    for j in range(2, b+1):
        if i % j == 0:
            cnt += 1
        if cnt ==2 :
            break
    if cnt == 1 :
        print(i)
```

```python
def isPrime(num):
    if num == 1:
        return False
    else:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True
 
 
M, N = map(int, input().split())
 
for i in range(M, N + 1):
    if isPrime(i):
        print(i)
```

```python
## 잘못된 코드 2,3 을 출력못함!
def prime_list(n):
	remove = [True] * n
	
	m = int(n ** 0.5)
	for i in range(2, m+1):
		if remove[i] == True:
			for j in range(i+i, n, i):
				remove[j] = False
	return [i for i in range(2,n) if remove[i] == True]
prime_list(1)
```

```python
# 보완한 완전한 코드
def prime_list(n):
    prime = [True] * n
    
    m = int(n**0.5)
    if n <= 3:
        prime[0] = False
        return [i+1 for i in range(n) if prime[i]==True]
    else :     
        for i in range(2, m+1):
            if prime[i] == True:
                for j in range(i+i, n, i):
                    prime[j] = False
        return [i for i in range(2,n) if prime[i] == True]
prime_list(13)
```

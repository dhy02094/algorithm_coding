```python
def prime_list(n):
    prime = [True] * (n+1)
    
    m = int(n**0.5)
    if n <= 3:
        prime[0] = False   # f t t t
        return [i+1 for i in range(n) if prime[i]==True ]
    else:
        for i in range(2,m+1):
            if prime[i] == True:
                for j in range(i+i,n+1,i):
                    prime[j]=False
        return [i for i in range(2,n+1) if prime[i]==True]  
            
t = int(input())

for _ in range(t):
    a = int(input())
    b = prime_list(a)
    hubo = []
    d = []
    for i in b:
        for j in b:
            if i+j == a:
                if i <= j :
                    hubo.append([i,j])
                    d.append(j-i)
    ans = hubo[d.index(min(d))]
    print(ans[0], ans[1])
                    
        
    
```

```python
prime_list = [False, False] + [True]*10002

for i in range(2,10002):
    if prime_list[i]:
        for j in range(i+i,10002,i):
            prime_list[j] = False
            
T = int(input())

for i in range(T):
    n = int(input())
    a = n//2
    b = a
    while 1:
        if prime_list[a] and prime_list[b]:
            print(a,b)
            break
        else:
            a-=1
            b+=1
```

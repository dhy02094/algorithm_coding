```python
T = int(input())
for _ in range(T):
    x1,y1,r1,x2,y2,r2 = map(int,input().split())
    d = ((x2 - x1) ** 2 + (y2 - y1) ** 2) **(1/2)
    if d == 0 and r1 == r2 :
        print(-1)
    elif abs(r1-r2) <d < r1 + r2 :
        print(2)
    elif d == r1+r2 or abs(r1-r2) == d :
        print(1)
    else :
        print(0)
```

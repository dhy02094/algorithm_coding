```python
def star(n):
    print('***' * (n//3))
    print('* *' * (n//3))
    print('***' * (n//3))
    
    return star(n)

star(27)
```

```python
def star(n):
    if n == 1:
        return ['*']
    
    stars = star(n//3)
    a = []
    
    for i in stars:
        a.append(i*3)
    for i in stars:
        a.append(i+' '* (n//3) +i)
    for i in stars:
        a.append(i*3)
    return a

n = int(input())
print('\n'.join(star(n)))
```

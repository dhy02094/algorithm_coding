```python
n = int(input())

a = list(map(int, input().split()))

inc = [1 for i in range(n)]

for i in range(n):
    for j in range(i):
        if a[i] > a[j]:
            inc[i] = max(inc[i], inc[j]+1)
            
dec = [1 for i in range(n)]

for i in range(n-1, -1, -1):
    for j in range(n-1,i,-1):
        if a[i] > a[j] :
            dec[i] = max(dec[i], dec[j]+1)
            
rlt = [0 for i in range(n)]

for i in range(n):
    rlt[i] = inc[i]+dec[i] - 1
    
print(max(rlt))
            
```

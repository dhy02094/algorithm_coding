```python
n = int(input())

drink = [0] * 10000
for i in range(n):
    drink[i] = int(input())
    
vol = [0]*10000
vol[0]=drink[0]
vol[1]=drink[1] + drink[0]
vol[2]=max(drink[0]+drink[2], drink[1]+drink[2], vol[1])
for i in range(3,n):
    vol[i] = max(vol[i-2]+drink[i], drink[i-1]+drink[i]+vol[i-3], vol[i-1])

print(max(vol))
```

```python
a = int(input())
b = list(map(int, input().split()))

c = [1] * a

for i in range(a):                    # 10
    for j in range(i):
        if b[j] < b[i] :              
            c[i] = max(c[i], c[j]+1) 
print(max(c))
    
    
```

```python
n, m = map(int, input().split())
val = []
for _ in range(n):
    val.append(int(input()))
val.sort(reverse=True)
cnt = 0
for i in val:
    if m % i == 0 :
        cnt += m//i
        break
    else :
        cnt += m//i
        m = m % i
print(cnt)
    
```

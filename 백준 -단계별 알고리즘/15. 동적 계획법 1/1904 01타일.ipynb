```python
# 1.    
import sys 
input = sys.stdin.readline

a = [0,1,2]

def tile(num):
    length = len(a)
    if num >= length: 
        for i in range(length,num+1):
            a.append(a[i-2]+a[i-1])
    return a[num]

tile(int(input()))
```



https://www.acmicpc.net/problem/15652

```python
a, b = map(int, input().split())
ans = []

def bts(start):
    if len(ans) == b :
        print(' '.join(map(str,ans)))
        return
    
    for i in range(start, a+1):
        ans.append(i)
        bts(i)
        ans.pop()
        start += 1
        
        
bts(1)
        
    
```

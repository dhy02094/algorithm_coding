```python
sudoqu = []
blank = []
for _ in range(9):
    a = list(map(int,input().split()))
    sudoqu.append(a)

for i in range(9):
    for j in range(9):
        if sudoqu[i][j] == 0:
            blank.append((i,j))

def rule_row(x,n):
    for i in range(9):
        if n == sudoqu[x][i]:
            return False
    return True

def rule_col(y,n):
    for i in range(9):
        if n == sudoqu[i][y]:
            return False
    return True

def rule_33(x,y,n):
    nx = x//3 * 3
    ny = y//3 * 3
    for i in range(3):
        for j in range(3):
            if n == sudoqu[nx+i][ny+j]:
                return False
    return True

def bt(idx):
    if idx == len(blank):
        for i in range(9):
            print(*sudoqu[i])
        exit(0)
        
        
    for i in range(1,10):
        x = blank[idx][0]
        y = blank[idx][1]
        
        if rule_row(x,i) and rule_col(y,i) and rule_33(x,y,i):
            sudoqu[x][y] = i
            bt(idx+1)
            sudoqu[x][y] = 0
            
bt(0)
```

https://www.acmicpc.net/problem/15651



```python
a ,b = map(int, input().split())
ans = []

def bts():
    if len(ans) == b:
        print(' '.join(map(str,ans)))
        return
    
    for i in range(1, a+1):
        ans.append(i)
        bts()
        ans.pop()
            
            
bts()
        
```

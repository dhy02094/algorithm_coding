```python
# 오류 왜 오류인지 못 찾았다.

N = int(input())
result = 0
b = 0
while N != b :
    if N < 10 :
        b = 10 * N + N
        result += 1
    else :
        c = int(N/10)
        d = N%10
        e = c+d
        f = e%10
        b = d * 10 + f
        result += 1
        


```

```python
N = int(input())
result = 0
num = N
while 1 :
    sum_num = int(num/10) + num%10
    new_num = (num%10)*10 + (sum_num % 10)
    result += 1
    
    if new_num == N :
        break
    num = new_num
print(result)
```

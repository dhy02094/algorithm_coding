```python
from collections import Counter
import sys

n = int(sys.stdin.readline())
stat = []
for _ in range(n):
    stat.append(int(sys.stdin.readline()))


a = sum(stat)/len(stat)
print(int(round(a,0)))

#2
stat.sort()
b = len(stat)//2
print(stat[b])

#3
cnt = Counter(stat).most_common(2)

if len(stat) > 1:
    if cnt[0][1] == cnt[1][1]:
        print(cnt[1][0])
    else:
        print(cnt[0][0])
else:
    print(cnt[0][0])

#4
d = stat[-1] - stat[0]
print(d)



    
```

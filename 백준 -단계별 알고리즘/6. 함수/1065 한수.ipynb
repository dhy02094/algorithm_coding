```python
def solve(n):
    clt = 0
    if n < 100 :
        clt = n
        return(clt)
    else : 
        for i in range(100,n+1):
            b = list(str(i))
            if int(b[1]) - int(b[0]) == int(b[2]) - int(b[1]):
                clt += 1
        return(clt + 99)
            
a = int(input())
print(solve(a))
        
        
```

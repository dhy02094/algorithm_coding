```python
n, k = map(int, input().split())
base = []
for _ in range(n):
    base.append(int(input()))
base.sort(reverse=True)
cnt = 0
for i in base:
    if i <= k:
        cnt += k//i
        k -= i*(k//i)
print(cnt)
```

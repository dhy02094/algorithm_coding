```python
n = int(input())

time = []
for _ in range(n):
    time.append(list(map(int, input().split())))
    
time = sorted(time, key=lambda x : x[0])
time = sorted(time, key=lambda x : x[1])

last = 0
cnt = 0

for i,j in time:
    if i >= last :
        cnt+= 1
        last =j
print(cnt)
```

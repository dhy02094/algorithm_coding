```python
# 내풀이

C = int(input())
for i in range(C):
    a = list(map(int,input().split()))
    b = a[0]
    c = a[1:]
    d = sum(c)/b
    e = []
    for i in c :
        if i > d:
            e.append(i)
    ans = len(e)/b * 100
    print(f'{ans:.3f}%')
    
    
        
        
```

```python
# 다른 사람 풀이

n = int(input())

for _ in range(n):
    nums = list(map(int, input().split()))
    avg = sum(nums[1:])/nums[0]  # 평균을 구함 (nums[0]: 학생수, nums[1:] 점수)
    cnt = 0
    for score in nums[1:]:
        if score > avg:
            cnt += 1  # 평균 이상인 학생 수
    rate = cnt/nums[0] *100
    print(f'{rate:.3f}%')
```
